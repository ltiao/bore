```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
def f(num_train, batch_size, num_steps):

    steps_per_epoch = np.ceil(np.true_divide(num_train, batch_size))
    num_epochs = num_steps // steps_per_epoch

    return num_epochs
```

```python
log_batch_sizes = np.arange(3, 7)
batch_sizes = 2**log_batch_sizes
X_grid = np.arange(1, 2000)
```

```python
frames = []

for batch_size in batch_sizes:

    frames.append(pd.DataFrame(dict(batch_size=batch_size, 
                                    dataset_size=X_grid, 
                                    num_epochs=f(X_grid, batch_size=batch_size, num_steps=1000))))

data = pd.concat(frames, axis="index", sort=True)
```

```python
fig, ax = plt.subplots()

sns.lineplot(x="dataset_size", y="num_epochs", hue="batch_size", palette="deep", data=data, ax=ax)

plt.show()
```

```python
fig, ax = plt.subplots()

sns.lineplot(x="dataset_size", y="num_epochs", hue="batch_size", palette="deep", data=data, ax=ax)

ax.set_xscale("log")
ax.set_yscale("log")

plt.show()
```
